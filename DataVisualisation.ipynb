```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/ToyotaCorolla.csv')
```

```python
data
```

```python
data.dropna(axis=1,inplace=True)
```

```python
data
```

```python
plt.scatter(data["Age"],data["Price"],c="red")
plt.xlabel("Age")
plt.ylabel("Price")
plt.show()
```

```python
data.describe()
```

```python
plt.hist(data['FuelType'],color='red',edgecolor='white',bins=5)
plt.show()
```

```python
counts=[979,800,678]
fuelType=['Diesel','Petrol','CNG']
index=np.arange(len(fuelType))
plt.bar(index,counts,color=['red','blue','cyan'])
plt.xlabel()
plt.show()
```

```python
plt.scatter(data["Age"],data["Price"],c=data["MetColor"],s=data['MetColor'])
plt.xlabel("Age")
plt.ylabel("Price")
plt.colorbar()
plt.show()
```
